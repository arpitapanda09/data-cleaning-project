# ==============================

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# LOAD DATASET
# ------------------------------

# Read CSV File
data = pd.read_csv("student_data.csv")

# Remove Extra Spaces from Column Names
data.columns = data.columns.str.strip()

# Display Original Dataset
print("Original Dataset:\n")
print(data)

# ------------------------------
# DATA CLEANING
# ------------------------------

# Check Missing Values
print("\nMissing Values:\n")
print(data.isnull().sum())

# Convert Age and Marks to Numeric
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Marks'] = pd.to_numeric(data['Marks'], errors='coerce')

# Fill Missing Age with Mean
data['Age'] = data['Age'].fillna(data['Age'].mean())

# Fill Missing Marks with Mean
data['Marks'] = data['Marks'].fillna(data['Marks'].mean())

# Fill Missing City with "Unknown"
data['City'] = data['City'].fillna("Unknown")

# Remove Duplicate Rows
data = data.drop_duplicates()

# ------------------------------
# CLEANED DATA
# ------------------------------

print("\nCleaned Dataset:\n")
print(data)

# Save Cleaned Dataset
data.to_csv("cleaned_student_data.csv", index=False)

# ------------------------------
# REPORT GENERATION
# ------------------------------

print("\nDataset Summary:\n")
print(data.describe())

# ------------------------------
# VISUALIZATION
# ------------------------------

plt.figure(figsize=(8, 5))

plt.bar(data['Name'], data['Marks'])

plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student Marks Report")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()

# ------------------------------
# SUCCESS MESSAGE
# ------------------------------

print("\nData Cleaning and Reporting Completed Successfully!")

# ==============================
# END PROJECT
# ==============================
```
